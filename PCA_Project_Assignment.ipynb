```python
df_final = df.drop(['OpenPorchSF','MasVnrArea','LotArea'], axis=1)
```

```python
df_final.columns
```

```python
X=df_final.drop(["SalePrice"],axis=1)
y=df_final["SalePrice"]
```

## Scaling data

```python
from sklearn.preprocessing import MinMaxScaler

scaled = MinMaxScaler()
X_scaled = scaled.fit_transform(X)
```

##PCA

```python
# import PCA from decomposition
from sklearn.decomposition import PCA

# create the PCA object to use to fit the given data
pca = PCA()

# fitting the PCA with the cleaned-up data
pca.fit(X_scaled)
```

```python
# Principal axes in feature space, representing the directions of maximum variance in the data.
print(pca.components_)
```

```python
# check the shape 
print(pca.components_.shape)
```

```python
# check the type 
print(type(pca.components_))
```

```python
# Percentage of variance explained by each of the selected components
print(pca.explained_variance_)
```

```python
print(len(pca.explained_variance_))
```

```python
print(pca.singular_values_)
```

```python
print(pca.mean_)
```

```python
print(pca.noise_variance_)
```

## Choosing the Number of Principal Components

```python
principal_components = pca.fit_transform(X_scaled)

print(type(principal_components))
print(principal_components.shape)

principal_components_df = pd.DataFrame(principal_components)
```

```python
sns.heatmap(principal_components_df.corr())
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

15 components contain approximated 90% of the variance. 

##PCA for Dimensionality Reduction

```python
pca_low_dim = PCA(n_components = 15)
```

```python
pca_low_dim.fit_transform(X)
```

```python
type(pca_low_dim)
```

```python
# original components
print(X_scaled.shape)

# transformed components
X_pca =pca_low_dim.transform(X_scaled)
print(X_pca.shape)
```

```python
print(pca_low_dim.explained_variance_)
```

```python
print(pca_low_dim.explained_variance_.sum())
```

##Linear Regression

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_pca,y, test_size=0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

ln=LinearRegression()
ln.fit(X_train,y_train)
```

```python
y_pred=ln.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

print(mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
print("Accuracy of model is : ",ln.score(X_test,y_test))
```
